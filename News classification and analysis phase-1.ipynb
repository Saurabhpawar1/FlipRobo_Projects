```python
#importing all the required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException,InvalidArgumentException,InvalidCookieDomainException
```

```python
#connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\saurabh\Downloads\chromedriver_win32_1\chromedriver.exe")
```

```python
#specifying the url of the webpage to be scraped
url="https://www.ebmnews.com/page/1047/"
```

```python
#opening the webpage through our driver
driver.get(url)
```

```python
#creating empty lists
Vertical=[]
Headlines=[]
Date=[]
Author=[]
Description=[]
```

```python
#getting the list of url of news
URL=[]
for i in range(1047,1082):
    news=driver.find_elements_by_xpath("//h2[@class='title']")
    for i in news:
        news1=i.find_element_by_tag_name("a")
        URL.append(news1.get_attribute('href'))
    driver.find_element_by_xpath("//a[@class='next page-numbers']").click()
    time.sleep(3)
```

```python
URL
```

```python
#going to the url of first news through our driver
driver.get(URL[0])
```

```python
#running a loop to extract all the required information from the list_of_url
from tqdm import tqdm
for i in tqdm(URL):
    driver.get(i)
    try:
        vertical=driver.find_element_by_xpath('//div[@class="term-badges "]/span/a').text
        Vertical.append(vertical)
    except NoSuchElementException:
        Vertical.append("Null")
    try:
        headlines=driver.find_element_by_xpath('//span[@class="post-title"]').text
        Headlines.append(headlines)
    except NoSuchElementException:
        Headlines.append("Null")

    try:
        description=driver.find_element_by_xpath("//div[@class='entry-content clearfix single-post-content']").text.replace("\n"," ")
        Description.append(description)
    except NoSuchElementException:
        Description.append("Null")
    time.sleep(3)
```

```python
#going to the webpage through our driver
driver.get(url)
```

```python
#scraping Date and Author
for i in range(1047,1082):
    for i in driver.find_elements_by_xpath("//time[@class='post-published updated']"):
        Date.append(i.text)
        
    for i in driver.find_elements_by_xpath("//i[@class='post-author author']"):
        Author.append(i.text)
        
    driver.find_element_by_xpath("//a[@class='next page-numbers']").click()
    time.sleep(3)
```

```python
#creating a dataframe

EBM_NEWS=pd.DataFrame({})
EBM_NEWS['Vertical']=Vertical
EBM_NEWS['Headlines']=Headlines
EBM_NEWS['Date']=Date
EBM_NEWS['Author']=Author
EBM_NEWS['Description']=Description
```

```python
#converting into csv format
EBM_NEWS.to_csv("EBM_NEWS_csv.csv")
```

```python
#converting into json format
EBM_NEWS.to_json (r'C:\Users\saurabh\OneDrive\Desktop\New folder (3)\EBM_NEWS_json.json')
```

```python
#converting into xml format

def to_xml(EBM_NEWS, filename=None, mode='w'):
    def row_to_xml(row):
        xml = ['<item>']
        for i, col_name in enumerate(row.index):
            xml.append('<field name="{0}">{1}</field>'.format(col_name, row.iloc[i]))
        xml.append('</item>')
        return '\n'.join(xml)
    res = '\n'.join(EBM_NEWS.apply(row_to_xml, axis=1))

    if filename is None:
        return res
    with open(filename, mode, encoding="utf8") as f:
        f.write(res)

pd.DataFrame.to_xml = to_xml
```

```python
EBM_NEWS.to_xml('EBM_NEWS_xml.xml')
```
