# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
page=requests.get("https://www.imdb.com/list/ls091520106/")
```

```python
page
```

```python
page.content
```

```python
bs=BeautifulSoup(page.content, 'html.parser')
```

```python
Movie_title=[]
Rating=[]
Year_release=[]
```

```python
titles=bs.find_all('div',class_="lister-item mode-detail")
titles[0:4]
```

```python
for i in titles:
    
    name=i.h3.a.text
    Movie_title.append(name)
Movie_title
```

```python
years=bs.find_all('span',class_="lister-item-year text-muted unbold")
years[0:4]
```

```python
for i in years:
    Year_release.append(i.get_text())
Year_release
```

```python
Release_year = list(map(lambda sub:int(''.join( 
      [ele for ele in sub if ele.isnumeric()])), Year_release))
Release_year
```

```python
ratings=bs.find_all('div',class_="ipl-rating-star small")
ratings[0:4]
```

```python
for i in ratings:
    Rating.append(i.get_text().replace("\n",""))
Rating
```

```python
print(len(Movie_title),len(Release_year),len(Rating))
```

```python
import pandas as pd
Movies=pd.DataFrame({})
Movies['Movie_title']=Movie_title
Movies['Rating']=Rating
Movies['Year']=Release_year
```

```python
Movies
```

```python
Movies.to_csv("Top100 IMDB movie ratings.csv")
```

# 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=8a7876cd-2844-4017-846a-2c0876945b7b&pf_rd_r=Z631HTQ28A104299NQH8&pf_rd_s=right-5&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_india_tr_rhs_1")
```

```python
page
```

```python
page.content
```

```python
Year = list(map(lambda sub:int(''.join( 
      [ele for ele in sub if ele.isnumeric()])), Year_release))
Year
```

```python
ratings=bs.find_all('td',class_="ratingColumn imdbRating")[0:100]
ratings[0:4]
```

```python
for i in ratings:
    Rating.append(i.get_text().replace("\n",""))
Rating
```

```python
print(len(Movie_title),len(Year),len(Rating))
```

```python
Movies=pd.DataFrame({})
Movies['Movie_title']=Movie_title
Movies['Rating']=Rating
Movies['Year']=Year
```

```python
Movies
```

```python
Movies.to_csv("Top100 IMDB indian movie ratings.csv")
```

# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

## Book 1

```python
page=requests.get("https://bookpage.com/reviews/25848-leuyen-pham-outside-inside-childrens#.X_iT0ugzZPY")
```

```python
page
```

```python
page.content
```

```python
bs=BeautifulSoup(page.content, 'html.parser')
```

```python
Book_title=[]
Author_name=[]
Genre=[]
Book_review=[]
```

```python
title = bs.find_all('h1',class_='italic')
title
```

```python
for i in title:
    
    Book_title.append(i.get_text().replace("\n",""))
Book_title
```

```python
author=bs.find_all('h4',class_="sans")
author
```

```python
for i in author:
    Author_name.append(i.get_text().replace("\n",""))
Author_name
```

```python
genre=bs.find_all('p',class_="genre-links")
genre
```

```python
for i in genre:
    Genre.append(i.get_text().replace("\n",""))
Genre
```

```python
review=bs.find_all('div',class_="article-body")
review
```

```python
for i in review:
    Book_review.append(i.get_text().replace("\n",""))
Book_review
```

```python
import itertools

list1 = [[Book_title], [Author_name], [Genre], [Book_review]]
Book1 = list(itertools.chain(*list1))
```

```python
Book1
```
