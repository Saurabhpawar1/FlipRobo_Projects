```python
#importing all the required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException,InvalidArgumentException,InvalidCookieDomainException
```

```python
#connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\saurabh\Downloads\chromedriver_win32_1\chromedriver.exe")
```

```python
#specifying the url of the webpage to be scraped
url="https://timesofindia.indiatimes.com/archive/year-2018,month-12.cms"
```

```python
#opening the webpage through our driver
driver.get(url)
```

```python
#creating empty lists
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
```

```python
#clicking on the 1st date of dec 2018
driver.find_element_by_xpath('//*[@id="calender"]/tbody/tr[2]/td[7]/a').click()
```

```python
#getting the list of url

URL1=[]
news=driver.find_elements_by_xpath("//span[@style='font-family:arial ;font-size:12;color: #006699']/a")
for i in news:
    URL1.append(i.get_attribute('href'))
URL1
```

```python
len(URL1)
```

```python
#going to the url of first news through our driver
driver.get(URL1[0])
```

```python
#running a loop to extract all the required information from the list_of_url
from tqdm import tqdm
for i in tqdm(URL1):
    driver.get(i)
    
    try:
        date=driver.find_element_by_xpath("//div[@data-plugin='dateformat']").text
        Date.append(date)
    except NoSuchElementException:
        pass
        try:    
            date1=driver.find_element_by_xpath("//div[@class='_3Mkg- byline']").text
            Date.append(date1)
        except NoSuchElementException:
            Date.append("-")
        
    try:
        author=driver.find_element_by_xpath("//a[@class='auth_detail']").text
        Author.append(author)
    except NoSuchElementException:
        pass
        try:
            author1=driver.find_element_by_xpath("//a[@rel='noopener']").text
            Author.append(author1)
        except NoSuchElementException:
            Author.append("-")
    
    try:
        vertical=driver.find_element_by_xpath('//*[@id="container"]/div[16]/div/div/ol/li[2]/a/span').text
        Vertical.append(vertical)
    except NoSuchElementException:
        pass
        try:
            vertical1=driver.find_element_by_xpath('//*[@id="l2nav"]/div[1]/div[3]/ul/li[2]/a').text
            Vertical.append(vertical1)
        except NoSuchElementException:
            Vertical.append("-")
        
    try:
        headline=driver.find_element_by_xpath("//h1[@class='_23498']").text
        Headline.append(headline)
    except NoSuchElementException:
        pass
        try:
            headline1=driver.find_element_by_xpath("//h1[@class='heading1']").text
            Headline.append(headline1)
        except NoSuchElementException:
            Headline.append("-")

    try:
        description=driver.find_element_by_xpath("//div[@class='Normal']").text.replace("\n"," ")
        Description.append(description)
    except NoSuchElementException:
        pass
        try:
            description=driver.find_element_by_xpath("//div[@class='ga-headlines']").text.replace("\n"," ")
            Description.append(description)
        except NoSuchElementException:
            Description.append("-")
    time.sleep(3)
```
