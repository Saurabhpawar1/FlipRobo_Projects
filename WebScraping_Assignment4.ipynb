```python
#opening the webpage through our driver
driver.get(url)
```

```python
#clicking on recruiters page
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="root"]/div[1]/div/ul[1]/li[2]/a/div'))).click()
```

```python
#switching to new tab
driver.switch_to.window(driver.window_handles[1])
```

```python
#going to Data science recruiters page
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="skill"]/div[1]/div[2]/input'))).send_keys("Data science")
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="qsbFormBtn"]'))).click()
```

```python
#creating empty lists

naukri = {}
naukri['Name'] = []
naukri['Designation'] = []
naukri['Company'] = []
naukri['Skills'] = []
naukri['Location'] = []
```

```python
#running a loop to extract all the required information from all pages

for i in range(0,118):
    
    # Name
    name = driver.find_elements_by_xpath('//span[@class="fl ellipsis"]'.format(i))
    for n in name:
        naukri['Name'].append(n.text)
    

    # Designation
    designation = driver.find_elements_by_xpath('//span[@class="ellipsis clr"]'.format(i))
    for d in designation:
        naukri['Designation'].append(d.text)
    
    # Company
    company = driver.find_elements_by_xpath('//*[@id="tabP-1"]/div[1]/div/div[1]/div[1]/p/a[2]'.format(i))
    for c in company:
        naukri['Company'].append(c.text)
    
    # Skills
    skills = driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]'.format(i))
    for s in skills:
        naukri['Skills'].append(s.text)

    # Location
    loc = driver.find_elements_by_xpath('//small[@class="ellipsis"]'.format(i))
    for l in loc:
        naukri['Location'].append(l.text)
    
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//button[text()="Next"]'.format(i)))).click()    
```

### 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/

```python
#connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\saurabh\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
#specifying the url of the webpage to be scraped
url=" https://archive.ics.uci.edu/"
```

```python
#opening the webpage through our driver
driver.get(url)
```

```python
#creating empty lists

Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
#getting the list of url of Datasets
Dataset_URL=[]


dataset=driver.find_elements_by_xpath('/html/body/table[3]/tbody/tr/td/table/tbody/tr/td[2]/table/tbody/tr/td[2]/span')
for i in dataset:
    dataset1=i.find_element_by_tag_name("a")
    Dataset_URL.append(dataset1.get_attribute('href'))
Dataset_URL
```

```python
len(Dataset_URL)
```

```python
#going to the url of first phone through our driver
driver.get(Dataset_URL[0])
```

```python
from tqdm import tqdm
for i in tqdm(Dataset_URL):
    driver.get(i)
    
    
    try:
        name=driver.find_element_by_css_selector('body > table:nth-child(3) > tbody > tr > td > table:nth-child(1) > tbody > tr > td:nth-child(1) > p:nth-child(1) > span.heading > b').text
        Dataset_name.append(name)
    except NoSuchElementException:
        Dataset_name.append("-")
    try:
        datatype=driver.find_element_by_css_selector('body > table:nth-child(3) > tbody > tr > td > table:nth-child(2) > tbody > tr:nth-child(1) > td:nth-child(2) > p').text
        Data_type.append(datatype)
    except NoSuchElementException:
        Data_type.append("-")
    try:
        task=driver.find_element_by_css_selector('body > table:nth-child(3) > tbody > tr > td > table:nth-child(2) > tbody > tr:nth-child(3) > td:nth-child(2) > p').text
        Task.append(task)
    except NoSuchElementException:
        Task.append("-")
        
    try:
        attribute=driver.find_element_by_css_selector('body > table:nth-child(3) > tbody > tr > td > table:nth-child(2) > tbody > tr:nth-child(2) > td:nth-child(2) > p').text
        Attribute_type.append(attribute)
    except NoSuchElementException:
        Attribute_type.append("-")
        
    try:
        instances=driver.find_element_by_css_selector('body > table:nth-child(3) > tbody > tr > td > table:nth-child(2) > tbody > tr:nth-child(1) > td:nth-child(4) > p').text
        No_of_instances.append(instances)
    except NoSuchElementException:
        No_of_instances.append("-")
        
    try:
        Attribute=driver.find_element_by_css_selector('body > table:nth-child(3) > tbody > tr > td > table:nth-child(2) > tbody > tr:nth-child(2) > td:nth-child(4) > p').text
        No_of_attribute.append(Attribute)
    except NoSuchElementException:
        No_of_attribute.append("-")
        
    try:
        year=driver.find_element_by_css_selector('body > table:nth-child(3) > tbody > tr > td > table:nth-child(2) > tbody > tr:nth-child(2) > td:nth-child(6) > p').text
        Year.append(year)
    except NoSuchElementException:
        Year.append("-")
```
